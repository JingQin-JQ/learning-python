```python
import urllib
# import json
# import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

```python
def find_links(base_url):
    '''
    find links of pages for cats in marketplaas
    input:
        base_url: the first page of query results of kitten in marketplaats
    output: 
        a list of links of pages of query results of kitten in marketplaats   
    '''
    address = base_url + '1' # get the first link
   
    uh = urllib.urlopen(address)
    data = uh.read()
#     print len(data)
    
    soup = BeautifulSoup(data)
    links = soup.find_all('span', id="pagination-pages", class_="pagination-pages")

    last_page_links = links[0]("span",class_="last" )
    if len(last_page_links) > 1:
        # do something
        print "Why length of lastPageLinks greater than 1???"  
        return None

    pages = []
    last = last_page_links[0].get_text()  
    for i in range(int(last)):
        pages.append(base_url + str(i+1))
   
    return pages


```

```python
def find_cats_in_pages(base_url):
    '''
    find links of cat in each pages in marketplaas
    input:
        base_url: the link of pages of query results of kitten in marketplaats
    output: 
        a list of individual cats in all pages of query results of kitten in marketplaats   
    '''
    page_links = find_links(base_url) # use function: find_links to get links for pages
    print len(page_links),"pages"
    all_links = []
    for page in page_links:   # get links from every pages,add to a list
        
        uh = urllib.urlopen(page)
        data = uh.read()
        
        soup = BeautifulSoup(data)
        links = soup.find_all('span', class_="thumb-placeholder-centered juiceless-link " )
        
        
        for link in links:
            cat_link = link.get('data-url')
            if cat_link != None:
                all_links.append(cat_link)
    
    return all_links,len(all_links) # return whole list of cats,and number of links

find_cats_in_pages("https://www.marktplaats.nl/z/dieren-en-toebehoren/\
katten-en-kittens-raskatten-korthaar/katten.html?query=katten&categoryId=407&currentPage=")

```
